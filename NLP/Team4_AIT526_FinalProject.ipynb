## 3. Working of the Project
This project uses a multi-model natural language processing pipeline to analyze sentiment, emotion, and topical trends in social media text data.  The main actions taken during implementation are listed below:
- **Data Loading** :The CSV file containing social media posts is where the dataset is loaded.  Columns with redundant information, like nameless indices, are removed. Consistency is maintained by handling missing text fields by substituting empty strings for nulls.
-  **Text Preprocessing**: Regular expressions are used to clean up raw text, eliminating unnecessary elements including URLs, mentions, hashtags, special characters, punctuation, digits, and whitespace.  Additionally, the text is changed to lowercase to standardize input at every level.
-  **Stopwords Removal and Lemmatization**: Common English stopwords (e.g., "the", "is", "and") are removed using NLTK‚Äôs stopwords list. Each word is then lemmatized using WordNetLemmatizer to reduce it to its base or dictionary form. This step improves the quality of text for downstream NLP tasks by reducing redundancy.
-  **Sentiment Analysis (BERT)**: A pre-trained BERT model is utilized to perform sentiment analysis using the Hugging Face Transformers library.  Each text entry is categorized by the model as either positive, negative, or neutral, and confidence scores are provided.  The contextual accuracy of sentiment recognition is guaranteed by this deep learning-based method.
-  **Emotion Detection (NRC Emotion Lexicon)**: A list of fundamental human emotions, including joy, sadness, fear, rage, and surprise, is mapped to terms in the text by the NRC Emotion Lexicon.  The dominating emotion for each document is determined by adding together the word-level emotion scores.  This offers emotional insights that go beyond polarity.
-  **Topic Modeling (BERTopic)**: The dataset's hidden themes and conversation points can be found using BERTopic.  It creates interpretable topic clusters by combining class-based TF-IDF with transformer embeddings.  The model provides important phrases for each topic and enables interactive visualizations for exploration.
-  **Visualization**: To display the findings in an understandable manner, a variety of plots and charts are created.  These include interactive BERTopic plots, emotion frequency histograms, sentiment distribution bar charts, and word clouds for topics and emotions.  These images facilitate both rapid interpretation and in-depth study.
-  **Error Handling and Robustness**: The pipeline incorporates checks to guarantee data integrity at every stage, handling for missing or distorted text, and exceptions during model inference.  These steps improve the project's reproducibility and dependability across various contexts and datasets.
- **VADER and BERT Comparison**: The results indicate that BERT outperforms VADER in terms of accuracy. BERT achieved an accuracy of 71%, while VADER's accuracy was only 43%. This suggests that BERT, being a transformer-based model, provides more reliable sentiment predictions compared to the rule-based VADER approach. This comparison underscores the advantage of deep learning models over rule-based systems when handling complex sentiment analysis tasks.


Comprehensive analysis of social media information is made possible by this modular and tiered approach, which meaningfully captures topic structures, emotional clues, and sentiments.

## 5.Implementation Process

### 1) Load the dataset

```python
import pandas as pd

dataframe = pd.read_csv('C:/Users/nehad/Desktop/Courses/SecondSem/AIT-NLP-526/Team4-project/sentimentdataset.csv')

# Remove any unnecessary columns
dataframe.drop(columns=[col for col in dataframe.columns if 'Unnamed' in col], inplace=True)

# Fill missing text
dataframe['Text'] = dataframe['Text'].fillna('')

dataframe.head()
```

### 2) Text preprocessing: Cleaning

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'@\w+|#\w+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

dataframe['clean_text'] = dataframe['Text'].apply(clean_text)
dataframe[['Text', 'clean_text']].head()
```

### 3) Text preprocessing: Lemmatization

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def lemmatize_text(text):
    tokens = nltk.word_tokenize(text)
    tokens = [lemmatizer.lemmatize(token.lower()) for token in tokens if token.lower() not in stop_words and token.isalpha()]
    return ' '.join(tokens)

dataframe['lemmatized_text'] = dataframe['clean_text'].apply(lemmatize_text)
dataframe[['clean_text', 'lemmatized_text']].head()
```

### 4)BERT Sentiment Classification

```python
from transformers import pipeline

# Make use of the BERT sentiment model
sentiment_model = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")

# Use the dataset (BERT input must to be less than 512 tokens).
dataframe['bert_sentiment'] = dataframe['lemmatized_text'].apply(lambda x: sentiment_model(x[:512])[0]['label'])
dataframe['bert_sentiment'] = dataframe['bert_sentiment'].map({'POSITIVE': 'Positive', 'NEGATIVE': 'Negative'})

dataframe[['Text', 'bert_sentiment']].head()
```

### 5) Error Analysis

```python
if 'Sentiment' in dataframe.columns:
    dataframe['original_sentiment'] = dataframe['Sentiment']
    error_samples = dataframe[dataframe['bert_sentiment'] != dataframe['original_sentiment']]
    display(error_samples[['Text', 'original_sentiment', 'bert_sentiment']].head(10))
```

### 6) Emotion Detection using NRCLex

```python
from nrclex import NRCLex

def get_emotions(text):
    emotion = NRCLex(text)
    return emotion.top_emotions

dataframe['emotions'] = dataframe['lemmatized_text'].apply(lambda x: get_emotions(x))
dataframe[['lemmatized_text', 'emotions']].head()
```

### 7) Visualization

#### 1. Word Count

```python
dataframe['word_count'] = dataframe['lemmatized_text'].apply(lambda x: len(x.split()))

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(dataframe['word_count'], bins=30, kde=True)
plt.title('Distribution of Word Count')
plt.xlabel('Number of Words')
plt.ylabel('Frequency')
plt.show()
```

#### 2. Sentiment Distribution (BERT) 

```python
sentiment_counts = dataframe['bert_sentiment'].value_counts()
sentiment_dataframe = sentiment_counts.reset_index()
sentiment_dataframe.columns = ['sentiment', 'count']

plt.figure(figsize=(6, 4))
sns.barplot(data=sentiment_dataframe, x='sentiment', y='count', hue='sentiment', palette='coolwarm', legend=False)
plt.title('Sentiment Distribution (BERT)')
plt.ylabel('Number of Posts')
plt.xlabel('Sentiment')
plt.tight_layout()
plt.show()
```

#### 3. Emotion Frequency Distribution

```python
from nrclex import NRCLex

def detect_emotion(text):
    emotion = NRCLex(text)
    if emotion.top_emotions:
        return emotion.top_emotions[0][0]
    else:
        return "neutral"

# Create the emotion column (if not exists)
dataframe['emotion'] = dataframe['Text'].apply(detect_emotion)

emotion_counts = dataframe['emotion'].value_counts()
emotion_dataframe = pd.DataFrame(list(emotion_counts.items()), columns=['emotion', 'count'])

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=emotion_dataframe, x='emotion', y='count', hue='emotion', palette='viridis', legend=False)
plt.title('Emotion Frequency Distribution (via NRCLex)')
plt.xlabel('Emotion')
plt.ylabel('Number of Posts')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### 4. Emotion vs Sentiment Heatmap

```python
dataframe['dominant_emotion'] = dataframe['emotions'].apply(
    lambda x: x[0][0] if isinstance(x, list) and len(x) > 0 else 'unknown'
)
# heatmap
heat_data = pd.crosstab(dataframe['bert_sentiment'], dataframe['dominant_emotion'])

plt.figure(figsize=(12,6))
sns.heatmap(heat_data, cmap='YlGnBu', annot=True, fmt='d')
plt.title('Emotion vs Sentiment Heatmap')
plt.ylabel('BERT Sentiment')
plt.xlabel('Dominant Emotion')
plt.tight_layout()
plt.show()
```

#### 5. NRC Emotion Word Cloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd

all_emotions = []

for emo_list in dataframe['emotions'].dropna():
    if isinstance(emo_list, list):
        for tup in emo_list:
            if isinstance(tup, tuple) and len(tup) == 2:
                all_emotions.append(tup[0])  # Extract emotion label

# frequency count
emotion_counts = pd.Series(all_emotions).value_counts().to_dict()

# Create a word cloud.
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(emotion_counts)


plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("NRC Emotion Word Cloud (via NRCLex)")
plt.show()
```
